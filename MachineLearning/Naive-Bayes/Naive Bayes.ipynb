```python
X=data["Message"]
y=data['Category']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.Message,data.Category)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
X_train_Count = cv.fit_transform(X_train.values)
X_test_Count = cv.transform(X_test)
X_train_Count.toarray()[:2]
```

```python
print(X_train_Count.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
model.fit(X_train_Count,y_train)
```

```python
emails=["CFL is pleased to announce a German language course  to enhance your academic and professional journey.",
        "Learning German opens doors to new cultures, traditions, and experiences.",
        "Germany offers tuition-free or low-cost education to international students."]
emails_count=cv.transform(emails)
model.predict(emails_count)
```

```python
model.score(X_test_Count,y_test)
```

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test,y_test)
```

```python
clf.predict(emails)
```
