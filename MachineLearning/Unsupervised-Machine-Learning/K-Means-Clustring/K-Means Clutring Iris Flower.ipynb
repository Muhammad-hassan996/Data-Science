```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv(r"C:\Users\hassa\OneDrive\Desktop\Iris.csv")
```

```python
df.head(5)
```

```python
df=df.drop(columns=['Id','Species'])
```

```python
df
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
ss
```

```python
scale=ss.fit_transform(df)
```

```python
scale
```

```python
sns.pairplot(df)
```

```python

plt.scatter(df['SepalLengthCm'], df['SepalWidthCm'],cmap='viridis')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
km=KMeans(n_clusters=3,)
predict=km.fit_predict(scale)
```

```python
predict
```

```python
df['Cluster']=predict
```

```python
df
```

```python
km.cluster_centers_
```

```python
# Creating a pair plot with the clusters
sns.pairplot(df, hue='Cluster', vars=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'], palette='viridis')
plt.suptitle('Pair Plot with Clusters - Iris Dataset', y=1.02)
plt.show()
```

```python

centroids = km.cluster_centers_
plt.scatter(scale[:, 0], scale[:, 1], c=df['Cluster'], cmap='viridis', marker='o', label='Data Points')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='o', label='Centroids')
plt.xlabel('Feature 1 (Standardized)')
plt.ylabel('Feature 2 (Standardized)')
plt.title('K-Means Clustering with Centroids')
plt.legend()
plt.show()
```

```python

sse = []
k_rng = range(1, 10)

for k in k_rng:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(df[['SepalLengthCm', 'SepalWidthCm']])
    sse.append(km.inertia_)
plt.plot(k_rng, sse)
plt.xlabel('Number of clusters (k)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.show()
```
